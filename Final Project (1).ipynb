># GENERAL ANALYSIS

># Growth Of Population

```python
country['Growth Of Population (%)']= country['Birthrate']-country['Deathrate']
```

```python
country
```

```python
plt.figure(figsize=(16,9))
sns.catplot( x='Region',y='Growth Of Population (%)', data=country,kind='point',aspect=3)
plt.title("Growth of Population in different Regions",size=20)
plt.grid()
plt.xticks(rotation=45)
```

```python
Country1['Employees']= (country['Population']*country['Literacy (%)'])/100 
Employees1=Country1['Employees'].round(decimals=0)
Country1['Educated Employees']=Employees1
Country1=Country1.drop(columns='Employees')
Country1['Labours']= Country1['Population']-Country1['Educated Employees']
```

```python
Country1
```

```python
def tag(value):
    if value > 4:
        return "Less Migrated"
    else :
        return "More Migrated"
```

```python
Country1['Migrants'] = Country1['    Net migration'].apply(tag)
```

```python
Country1
```

```python
sns.catplot( x='Region',y='    Net migration', data=country,kind='bar',aspect=3)
plt.title("Net Migration in Different Region",size=20)
plt.xticks(rotation=90)
plt.grid()
```

```python
Country1=Country1.drop(columns='Climate')
Country1=Country1.drop(columns='Coastline (coast/area ratio)')
Country1=Country1.drop(columns='Pop. Density (per sq. mi.)'
Country1=Country1.drop(columns='Other (%)')
Country1=Country1.drop(columns='Phones (per 1000)')
```

```python
Country1
```

```python
#Why communication is Important for various sector
Phone_usage=(Country1['Population']*Country1['Phones (per 1000)'])/1000
Phone_usage1=(Phone_usage*100)/Country1['Population']
Phone_usage.round(decimals=0)
Country1['Percentage of Phone_usage']=Phone_usage1
```

```python
Country4=Country1.sort_values(['Population'],ascending=False)
Country4.groupby(['Population','Country'])['Percentage of Phone_usage'].mean().reset_index()
```
